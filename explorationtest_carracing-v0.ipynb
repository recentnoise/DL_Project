## Exploration of observation and action spaces

`step()` returns four values:
- `observation (object)`: observation of the environment. Can vary between environments.
- `reward (float)`: reward achieved by the previous action. Can vary between environments.
- `done (bool)`: wether we finished and should `.reset()` the environment or not.
- `info (dict)`: diagnostic information useful for debugging.

Each timestep, the agent chooses an `action`, and the environment returns an `observation` and a `reward`.

**Note that `reset()` returns the initial observation**

```python
## Get the initial observation
observation = env.reset()
score_list = []
t = 0
score = 0
while True:
    action = env.action_space.sample()
    env.render()
    observation, reward, terminated, truncated, info, = env.step(action)
    score += reward
    done = truncated or terminated 
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
    t += 1
    score_list.append(score)
    print('Average score', np.mean(score_list))
    
env.close()
```

### Actions

This environment's actions consist of 3 `float32` numbers: 
- Steer $\in$ $[-1,1]$
- Gas $\in$ $[0,1]$
- Brake $\in$ $[0,1]$

They are defined with the `Box` class, which represents a possibly unbounded in $R^n$.

```python
print("Sample action -> Steer: {} | Gas: {} | Brake: {}".format(*env.action_space.sample()))
print("Action Space: ", env.action_space)
```

### Wrapper
We can put the image pre-processing into a `gym.Wrapper` to ease it's use.

```python
class CropObsWrapper(gym.ObservationWrapper):
    """
    We crop the bottom part of the image to remove the part that contains no useful information.
    """
    
    def __init__(self, env):
        super(CropObsWrapper, self).__init__(env)

    def observation(self, observation):
        return observation[:-12,:]
```

```python
env_w = CropObsWrapper(env)
```

```python
observation = env_w.reset()
t = 0
while True:
    env_w.render()
    action = env_w.action_space.sample()
    observation, reward, terminated, truncated, info = env_w.step(action)
    done = terminated or truncated
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
    t += 1
    
env_w.close()
```

```python
plt.imshow(observation)
plt.show()
```

```python
class CannyObsWrapper(gym.ObservationWrapper):
    """
    Apply a canny filter to the observations.
    """
    def __init__(self, env):
        super(CannyObsWrapper, self).__init__(env)

    def observation(self, observation):
        return cv2.Canny(observation, 100, 200, L2gradient=True)   
```

```python
env_c = CannyObsWrapper(env)
```

```python
observation = env_c.reset()
t = 0
while True:
    env_c.render()
    action = env_c.action_space.sample()
    observation, reward, terminated, truncated, info = env_c.step(action)
    done = terminated or truncated
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
    t += 1
    
env_c.close()
```

```python
env = gym.make("CarRacing-v2")

# Reset the environment to its initial state
observation = env.reset()

# Initialize variables to keep track of the average reward
num_steps = 0
total_reward = 0
discount_factor = 0.95 # Define the discount factor
future_steps = 5 # Number of future steps to consider

# Create a list to store the data for each iteration
iterations_data = []

# Start the loop to run the environment
done = False
while not done:
    env.render() # Render the environment
    action = env.action_space.sample() # Choose a random action from the action space
    observation, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated
    if done:
        print("Episode finished after {} timesteps".format(num_steps+1))
        break
    num_steps += 1 # Increment the number of steps
    total_reward += reward * (discount_factor ** num_steps) # Accumulate the discounted reward

    future_reward = 0 # Initialize the future reward
    for i in range(future_steps):
        future_observation, future_reward, terminated, truncated, info = env.step(action) # Perform the chosen action and observe the results
        done = terminated or truncated
        if done:
            print("Episode finished after {} timesteps".format(num_steps+1))
            break
        num_steps += 1
        future_reward += future_reward * (discount_factor ** (num_steps + i + 1)) # Accumulate the discounted reward

    # Store the data for this iteration
    iteration_data = {
        "S": observation,
        "A": action,
        "S'": observation,
        "R": reward,
        "Qf": total_reward + future_reward
    }
    iterations_data.append(iteration_data)

# Calculate and print the average reward
average_reward = total_reward / num_steps
print("Average reward:", average_reward)

# Close the environment when finished
env.close()
```
